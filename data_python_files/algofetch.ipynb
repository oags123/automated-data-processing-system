```python
def fetch_prices_get_df(time_interval, symbol, start_time_iso, start_time_count):
    # Step 1: Instantiate the FetchPrices class
    fetcher = FetchPrices(time_interval, symbol, start_time_iso, start_time_count)

    # Step 2: Call the run_all method
    df = fetcher.run_all()

    return df
```

```python
from datetime import datetime, timezone, timedelta

def get_current_utc_iso_minus_10min():
    """Return the current UTC time minus 10 minutes in ISO 8601 format without microseconds."""
    dt = datetime.now(timezone.utc) - timedelta(minutes=10)
    return dt.replace(microsecond=0).isoformat().replace("+00:00", "Z")

#print(get_current_utc_iso_minus_10min())
# Output: 2025-09-12T14:25:27Z
start_time_1m = get_current_utc_iso_minus_10min()
print(start_time_1m)
```
